```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="left")

# Display the data table for preview
combined_data.head()
```

```python
#Checking the number of mice
mouse_count = len(combined_data["Mouse ID"].unique())
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_ID = combined_data.duplicated(subset=["Mouse ID", "Timepoint"], keep = False)
duplicate_ID

duplicate_mouse_ID = combined_data.loc[duplicate_ID, "Mouse ID"].unique()
duplicate_mouse_ID
```

```python
# Optional: Get all the data for the duplicate mouse ID.
all_duplicates = combined_data.loc[combined_data["Mouse ID"] == duplicate_mouse_ID[0]]
all_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined = combined_data.drop(all_duplicates.index)
clean_combined.head()
```

```python
new_mouse_count = len(clean_combined["Mouse ID"].unique())
new_mouse_count
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_mean = clean_combined.groupby("Drug Regimen").mean(numeric_only = True)["Tumor Volume (mm3)"]
drug_mean

drug_median = clean_combined.groupby("Drug Regimen").median(numeric_only = True)["Tumor Volume (mm3)"]
drug_median

drug_var = clean_combined.groupby("Drug Regimen").var(numeric_only = True)["Tumor Volume (mm3)"]
drug_var

drug_std = clean_combined.groupby("Drug Regimen").std(numeric_only = True)["Tumor Volume (mm3)"]
drug_std

drug_sem = clean_combined.groupby("Drug Regimen").sem(numeric_only = True)["Tumor Volume (mm3)"]
drug_sem
# Assemble the resulting series into a single summary DataFrame.
drug_df = pd.DataFrame({"Mean Tumor Volume": drug_mean,
                        "Median Tumor Volume": drug_median,
                        "Tumor Volume Variance": drug_var,
                        "Tumor Volume Std. Dev.": drug_std,
                        "Tumor Volume Std. Err.": drug_sem})
drug_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

drug_df2 = clean_combined.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])
drug_df2
```

# Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

final_volume = clean_combined.groupby(["Drug Regimen", "Mouse ID"]).last()[["Timepoint", "Tumor Volume (mm3)"]]
final_tumor = final_volume.reset_index()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_names = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
total_volume = []
            
#Calculate the IQR and quantitatively determine if there are any potential outliers. 
for name in treatment_names:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_volume = final_tumor.loc[final_tumor["Drug Regimen"] == name, "Tumor Volume (mm3)"]
     # add subset
    total_volume.append(final_volume)

    #Creating Quartiles and Bounds
    quartiles = final_volume.quantile([.25,.5,.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq - lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    # Determine outliers using upper and lower bounds
    outliers = final_volume.loc[(final_volume < lower_bound) | (final_volume > upper_bound)]

    #Print Results
    print(f"{name} potential outliers: ")
    if outliers.count() > 0:
        for i in outliers.index:
            print(f'Outlier Final Tumor Volume (mm3) is {round(outliers[i], 2)}')
    else: print("No Potential Outliers")
    print('----------------------------------------------------------')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, final_tumor_plot = plt.subplots()
final_tumor_plot.boxplot(total_volume, labels=treatment_names, flierprops={"markerfacecolor" : "r", "markersize" : 10})

#Adding Titles and x/y label
final_tumor_plot.set_title("Final Tumor Volume \n(Four Treatment Regimens)", fontweight="bold")
final_tumor_plot.set_xlabel("Drug Regimen", fontweight="bold")
final_tumor_plot.set_ylabel("Final Tumor Volume (mm3)", fontweight="bold")

#Present Graph
plt.tight_layout()
plt.show()
```

# Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_volume = clean_combined.loc[clean_combined["Drug Regimen"] == "Capomulin"]

#Identify specific mouse
mouse_id = "l509"
mouse_df = capomulin_volume.loc[capomulin_volume["Mouse ID"] == mouse_id]

#Create Plot
capomulin_plot = plt.plot(mouse_df["Timepoint"], mouse_df["Tumor Volume (mm3)"], marker="o", color = "green")

#Create Title and x/y Labels
plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Timepoint (Days)", fontweight = "bold")
plt.ylabel("Tumor Volume (mm3)", fontweight = "bold")

#Display Plot
plt.tight_layout()
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
tumor_weight = capomulin_volume.groupby("Mouse ID").mean(numeric_only=True)
weight_avg = tumor_weight["Weight (g)"]
tumor_avg = tumor_weight["Tumor Volume (mm3)"]

#Generate the Plot
plt.scatter(weight_avg, tumor_avg)

#Add Title and x/y Labels
plt.title("Weight vs Average Tumor Volume \n(Capomulin regimen)", fontweight="bold")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

#Display Plot
plt.show()
```

# Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr_co = st.pearsonr(tumor_weight["Weight (g)"], tumor_weight["Tumor Volume (mm3)"]);
print(f"The correlation between mouse weight and average tumor volume is {round((corr_co)[0],2)} ")

#Regression Work
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(tumor_weight["Weight (g)"], tumor_weight["Tumor Volume (mm3)"])
regress_values = tumor_weight["Weight (g)"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Generate Plot
plt.scatter(tumor_weight["Weight (g)"], tumor_weight["Tumor Volume (mm3)"])
plt.plot(tumor_weight["Weight (g)"], regress_values, color = "purple")
plt.annotate(line_eq, xy=(18,38))
plt.title("Average Tumor Volume by Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()


```
